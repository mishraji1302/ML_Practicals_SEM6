```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Statistical Analysis
df = pd.DataFrame(X, columns=iris.feature_names)
print("Mean:\n", df.mean())
print("\nMedian:\n", df.median())
print("\nStandard Deviation:\n", df.std())

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Model before scaling
model = KNeighborsClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("\nAccuracy before scaling:", accuracy_score(y_test, y_pred))

# Min-Max Scaling
minmax = MinMaxScaler()
X_train_mm = minmax.fit_transform(X_train)
X_test_mm = minmax.transform(X_test)

model.fit(X_train_mm, y_train)
y_pred_mm = model.predict(X_test_mm)
print("Accuracy after Min-Max Scaling:", accuracy_score(y_test, y_pred_mm))

# Standardization
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)

model.fit(X_train_std, y_train)
y_pred_std = model.predict(X_test_std)
print("Accuracy after Standardization:", accuracy_score(y_test, y_pred_std))
```
